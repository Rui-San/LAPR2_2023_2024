```python
import csv
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def showBarplot(startMonth, endMonth, parkID, parkMatrix):
    fig, ax = plt.subplots()

    monthsDesc = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

    months = monthsDesc[startMonth-1:endMonth]
    
    values = [parkMatrix[i][2] for i in range(int(startMonth)-1, int(endMonth))]
    
    ax.bar(months, values)

    ax.set_ylabel('water consumption (m3)')
    ax.set_title(parkID + ' Park water consumption from ' + monthsDesc[startMonth-1] + ' to ' + monthsDesc[endMonth-1])

    plt.xticks(rotation='vertical')

    plt.show()
```

```python
def US009_opt1(grouped_matrix):
    parkID = input("What park would you like to see the water consumption for? ")
    while parkID not in grouped_matrix[:,0]:
        parkID = input("What park would you like to see the water consumption for? ")

    startMonth = int(input("Enter the start month (1-12): "))
    while startMonth < 1 or startMonth > 12:
        startMonth = int(input("Enter the start month (1-12): "))

    endMonth = int(input("Enter the end month (1-12): "))
    while endMonth <= startMonth or endMonth > 12:
        endMonth = int(input("Enter the end month (1-12): "))

    parkMatrix = grouped_matrix[grouped_matrix[:,0] == parkID]

    showBarplot(startMonth, endMonth, parkID, parkMatrix)

    return 0
```

```python
def US009_opt2(grouped_matrix_total):
    numberOfParks = int(input("How many parks would you like to see monthly costs for? "))
    while numberOfParks < 1:
        numberOfParks = int(input("How many parks would you like to see monthly costs for? "))
    park = list(range(numberOfParks))
    for i in range(0, numberOfParks):
        park[i] = input("Enter the "+ str(i) +" park ID: ")
        while park[i] not in grouped_matrix_total[:,0]:
            park[i] = input("Enter the "+ str(i) +" park ID: ")
    
    costUntil1000 = 0.7;
    extraAfter1000Percent = 0.15;

    for i in range(0, numberOfParks):
        index = np.where(grouped_matrix_total[:, 0] == park[i])[0]
        costCont = 0
        if grouped_matrix_total[index,1] > 1000:
            costCont = 1000 * costUntil1000;
            costExtra = (grouped_matrix_total[index,1] - 1000) * costUntil1000
            costCont = costCont + costExtra * extraAfter1000Percent + costExtra;
        else:
            costCont = grouped_matrix_total[i,1] * costUntil1000
        
        print("The cost for the " + grouped_matrix_total[index[0],0] + " park is: " + str(round(costCont[0], 2)) + "â‚¬")
```

```python
def US009_opt3(list_matrix, grouped_matrix_total):

    parkBiggerConsumption = grouped_matrix_total[np.argmax(grouped_matrix_total[:,1])]
    parkLeastConsumption = grouped_matrix_total[np.argmin(grouped_matrix_total[:,1])]

    onlyBiggestPark = list_matrix[list_matrix[:,0] == parkBiggerConsumption[0]]
    onlyLeastPark = list_matrix[list_matrix[:,0] == parkLeastConsumption[0]]

    biggerParkMean = np.mean(onlyBiggestPark[:,4])
    biggerParkMedian = np.median(onlyBiggestPark[:,4])
    biggerParkStandardDeviation = np.std(onlyBiggestPark[:,4])
    biggerParkCoeficientOfSkewness = (3 * (biggerParkMean - biggerParkMedian)) / biggerParkStandardDeviation

    print("The park with the biggest consumption is: " + parkBiggerConsumption[0] + " with a consumption of " + str(parkBiggerConsumption[1]) + "m3")
    print("- Mean consumption: " + str(biggerParkMean) + "m3")
    print("- Median consumption: " + str(biggerParkMedian) + "m3")
    print("- Standard deviation: " + str(biggerParkStandardDeviation))
    print("- Coeficient of skewness: " + str(biggerParkCoeficientOfSkewness))

    leastParkMean = np.mean(onlyLeastPark[:,4])
    leastParkMedian = np.median(onlyLeastPark[:,4])
    leastParkStandardDeviation = np.std(onlyLeastPark[:,4])
    leastParkCoeficientOfSkewness = (3 * (leastParkMean - leastParkMedian)) / leastParkStandardDeviation
    
    print("")
    print("The park with the least consumption is: " + parkLeastConsumption[0] + " with a consumption of " + str(parkLeastConsumption[1]) + "m3")
    print("- Mean consumption: " + str(leastParkMean) + "m3")
    print("- Median consumption: " + str(leastParkMedian) + "m3")
    print("- Standard deviation: " + str(leastParkStandardDeviation))
    print("- Coeficient of skewness: " + str(leastParkCoeficientOfSkewness))


    return 0
```

```python
# Define the name of the CSV file
file = "water_consumption.csv"

column_names = ['Park', 'Year', 'Month', 'Day', 'Consumption']

# Read the CSV file into a DataFrame    
df = pd.read_csv(file, names=column_names, sep=';', decimal=',', skiprows=1)

list_matrix = df.values
# Group the data by the first and third columns and sum the fourth column
grouped_df = df.groupby(["Park","Month"])["Consumption"].sum().reset_index()

# Convert the grouped DataFrame to a matrix
grouped_matrix = grouped_df.values
grouped_matrix_total = grouped_df.groupby("Park")["Consumption"].sum().reset_index().values


opt = -1
while opt != 0:
    opt = int(input("1. Barplot | 2. Average Cost | 3. Stat. Analysis | 0. Exit\n"))
    if opt == 1:
        US009_opt1(grouped_matrix)
    elif opt == 2:
        US009_opt2(grouped_matrix_total)
    elif opt == 3:
        US009_opt3(list_matrix, grouped_matrix_total)
```
